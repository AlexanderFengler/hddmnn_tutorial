```python
# Append system path to include the config scripts
import sys
import os
from time import time
from copy import deepcopy
import pickle
import numpy as np

#print('importing lanfactory')
#import lanfactory

#print('importing ssms')
#import ssms

print('importing hddm')
import hddm

sys.path.insert(1, os.path.join(sys.path[0], '..'))
#from config import *

import tensorflow
import torch

import pandas as pd
from hddm.simulators.hddm_dataset_generators import simulator_h_c
```

```python
def run_model(model_name = 'angle',
              network_path = '',
              network_config_path = '',
              model_config = '',
              parallel = 0,
              model_idx = 0,
              recov_idx = 0,
              custom = 1,
              include_regressor = 0):
     
    # MAKE DATA
    data, parameter_dict = simulator_h_c(data = None,
                                         n_subjects = 5,
                                         n_trials_per_subject = 200,
                                         model = model_name,
                                         p_outlier = 0.0,
                                         conditions = None,
                                         depends_on = None,
                                         regression_models = None,
                                         regression_covariates = None,
                                         group_only_regressors = True,
                                         group_only = None,
                                         fixed_at_default = None)
    
    print(data)
    
    data['covariate'] = np.random.normal(size = data.shape[0])
    
    # ASSIGN MODEL CONFIG
    #model_config = run_config['model_config']
    
    # LOAD NETWORK
    network_config = pickle.load(open(network_config_path, 'rb'))
    #network_path = run_config['lan_files'][model_idx]
    network = hddm.torch.mlp_inference_class.LoadTorchMLPInfer(model_file_path = network_path,
                                                               network_config = network_config,
                                                               input_dim = len(model_config["params"]) + 2)
    
    # PREP FOR REGRESSOR TEST
    reg_model_v = {'model': 'v ~ 1 + covariate', 'link_func': lambda x: x}
    reg_model_t = {'model': 't ~ 1', 'link_func': lambda x: x}
    reg_descr = [reg_model_t, reg_model_v]
    
    # ACTUAL RUN  ---------------------------------------------------
    for chain in range(0, 1, 1):
        start_t = time()
        if custom:
            if include_regressor:
                # Set up the regressor a regressor:
                print('RUNNING CUSTOM REGRESSOR')
                model_ = hddm.HDDMnnRegressor(data,
                                              reg_descr,
                                              #model = model_name,
                                              model_config = model_config,
                                              network = network,
                                              include = model_config["params"],
                                              group_only_regressors = True,
                                              is_group_model = True)
                db_file_name = 'data/test/db_model_custom_reg.db'
                model_file_name = 'data/test/model_custom_reg.pickle'
                
            else:
                print('RUNNING CUSTOM HDDMNN')
                model_ = hddm.HDDMnn(data,
                                     #model = model_name,
                                     model_config = model_config,
                                     network = network,
                                     include = model_config["params"],
                                     is_group_model = True)
                db_file_name = 'data/test/db_model_custom.db'
                model_file_name = 'data/test/model_custom.pickle'
                
        else:
            if include_regressor:
                print('RUNNING INCLUDED REGRESSOR')
                # Set up the regressor a regressor:
                model_ = hddm.HDDMnnRegressor(data,
                                              reg_descr,
                                              model = model_name,
                                              include = model_config["params"],
                                              is_group_model = True)
                db_file_name = 'data/test/db_model_reg.db'
                model_file_name = 'data/test/model_reg.pickle'
                
            else:
                print('RUNNING INCLUDED HDDMNN')
                model_ = hddm.HDDMnn(data,
                                     model = model_name,
                                     include = model_config["params"],
                                     is_group_model = True)
                db_file_name = 'data/test/db_model.db'
                model_file_name = 'data/test/model.pickle'

        # Check if model file name exists --> if yes don't run anything
        #if os.path.exists(model_file_name):
            #print(model_file_name, ' EXISTS --> NOT RUNNING THIS MODEL')
#         else:
#             #try:
        model_.sample(150, burn = 50, 
                      dbname = db_file_name, 
                       db = 'pickle')

        model_.save(model_file_name)

        print(model_.mc.step_method_dict)
#             except:
#                 print("Something went wrong with sampling from the model")
#                 return -1

    return 1
```

```python
network_path = 'data/torch_models/24595bd4a33e11eca6673cecef05799c_' + \
                    'training_data_angle_torch_state_dict.pt'
network_config_path = 'data/torch_models/24595bd4a33e11eca6673cecef05799c_' + \
                        'training_data_angle_torch__network_config.pickle'

my_config = deepcopy(hddm.model_config.model_config['angle'])
```

```python
for i in range(1):
    run_model(model_name = 'angle',
              network_path = network_path,
              network_config_path = network_config_path,
              model_config = my_config,
              parallel = 0,
              model_idx = 0,
              recov_idx = i,
              custom = 1,
              include_regressor = 1,
              group_only_regressor = 1)
```
