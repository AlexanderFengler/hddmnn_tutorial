```python
def conflict_detection_bound(a, shape, scale, theta, t, psi = 0):
    return np.maximum(0, a + scale*((shape*(t + psi)) / np.power(shape, (t + psi))) - theta*t)
```

```python
t_vec = np.arange(0, 10, 0.01)
bound_out = conflict_detection_bound(a = 2, shape = 1.6, scale = 1, theta = 0.0, t = t_vec, psi = 0)
```
